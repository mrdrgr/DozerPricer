## Getting Data Ready

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/TrainAndValid.csv', low_memory=False, parse_dates=['saledate'])
data.head(10)
```

```python
data.sort_values('saledate', inplace=True, ascending=True)
data.reset_index(drop=True, inplace=True)
data.head(10)
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data['SaleYear'] = data['saledate'].dt.year
data['SaleMonth'] = data['saledate'].dt.month
data['SaleDay'] = data['saledate'].dt.day
data['SaleDayOfTheWeek'] = data['saledate'].dt.dayofweek
data.T
```

```python
data.drop('saledate', axis=1, inplace=True)
data.T
```

## Converting Data To Numbers

```python
for label, content in data.items():
    if pd.api.types.is_string_dtype(content):
        data[label] = content.astype('category')
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
for label, content in data.items():
    if pd.api.types.is_numeric_dtype(content):
        data[label+'Missing'] = pd.isnull(content)
        data[label] = content.fillna(content.median())
```

```python
data.info()
```

```python
for label, content in data.items():
    if not pd.api.types.is_numeric_dtype(content):
        data[label+'Missing'] = pd.isnull(content)
        data[label] = pd.Categorical(content).codes + 1
```

```python
data.info(verbose=True)
```

```python
x_train = data[data['SaleYear'] != 2012].drop('SalePrice', axis=1)
x_test  = data[data['SaleYear'] == 2012].drop('SalePrice', axis=1)
y_train = data[data['SaleYear'] != 2012]['SalePrice']
y_test = data[data['SaleYear'] == 2012]['SalePrice']
```

## Create Prediction Model

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_squared_log_error, mean_absolute_error

def score_model(model, x, y):
    predictions = model.predict(x)
    r2 = r2_score(y, predictions)
    mae = mean_absolute_error(y, predictions)
    rmse = np.sqrt(mean_squared_error(y, predictions))
    rmsle = np.sqrt(mean_squared_log_error(y, predictions))
    return {'r2': r2, 'mae': mae, 'rmse': rmse, 'rmsle': rmsle}
```

```python
from sklearn.ensemble import RandomForestRegressor

base_model = RandomForestRegressor(n_jobs=-1, random_state=5)
base_model.fit(x_train, y_train)
score_model(base_model, x_test, y_test)
```

## Improving Model

```python
from sklearn.model_selection import RandomizedSearchCV

rs_model = RandomizedSearchCV(
    RandomForestRegressor(n_jobs=-1, random_state=5, max_samples=10000),
    {
        'n_estimators': np.arange(10, 250, 10),
        'max_depth': [None, 3, 5, 10, 25],
        'min_samples_split': np.arange(2, 20, 2),
        'min_samples_leaf': np.arange(1, 20, 2),
        'max_features': ['sqrt', 'log2', None],
    },
    n_iter=100,
    cv=5,
    verbose=True
)
rs_model.fit(x_train, y_train)
```

```python
rs_model.best_params_
```

```python
score_model(rs_model, x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

gs_model = GridSearchCV(
    RandomForestRegressor(n_jobs=-1, random_state=5, max_samples=10000),
    {
        'n_estimators': np.arange(50, 250, 50),
        'max_depth': [None, 5, 25],
        'min_samples_split': np.arange(2, 10, 2),
        'min_samples_leaf': np.arange(1, 20, 5),
        'max_features': ['sqrt', 'log2', None],    
    },
    cv=5,
    verbose=True
)
gs_model.fit(x_train, y_train)
```

```python
gs_model.best_params_
```

```python
score_model(gs_model, x_test, y_test)
```

```python
ideal_model = RandomForestRegressor(n_jobs=-1, random_state=5, n_estimators=150, min_samples_split=2, min_samples_leaf=1, max_features=None, max_depth=None)
ideal_model.fit(x_train, y_train)
score_model(ideal_model, x_test, y_test)
```

```python
feature_importance = pd.DataFrame({'Label': x_train.columns, 'Feature': ideal_model.feature_importances_})
feature_importance.sort_values('Feature', inplace=True, ascending=False)
feature_importance.set_index('Label', inplace=True)
feature_importance.head(10)
```
